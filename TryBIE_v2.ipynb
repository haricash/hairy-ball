```python
import numpy as np
```

```python
def G(xi,yi,xj,yj):
    r = np.array([[xi-xj, yi-yj]])
    rr = np.sqrt(r.dot(r.T))
    out = (np.eye(2)+np.outer(r,r)/rr**2)/rr
    return out
```

```python
N = 32
Np = 32
theta = np.linspace(0,2*np.pi,N)
x = np.cos(theta)
y = np.sin(theta)
```

```python
# build solution matrix

m = np.zeros((2*N,2*N))
for i in range(N):
    for j in range(N):
        if j != i:
            m[2*i:2*(i+1), 2*j:2*(j+1)] = G(x[i],y[i],x[j],y[j])
```

```python
# solve example flow field
v = np.zeros(2*N)
v[::2] = y*np.cos(theta)
v[1::2] = -x*np.cos(theta)
```

```python
from numpy.linalg import solve
f = solve(m, v)
```

```python
def get_v(xx, yy):
    v = np.zeros(2)
    for i in range(N):
        v += G(xx,yy,x[i],y[i]).dot(f[2*i:2*(i+1)])
    return v
```

```python
Np=32
_x, _y = np.meshgrid(np.linspace(-1.1,1.1,Np), np.linspace(-1.1,1.1,Np))
```

```python
import matplotlib.pyplot as plt
```

```python
vx = np.zeros_like(_x)
vy = np.zeros_like(_x)


for i in range(Np):
    for j in range(Np):
        vv = get_v(_x[i,j],_y[i,j])
        vx[i,j] = vv[0]
        vy[i,j] = vv[1]
```

```python
plt.quiver(x,y,v[::2],v[1::2],color='red')
plt.quiver(_x, _y, vx, vy)
```

```python
fig =plt.figure()
plt.streamplot(_x, _y, vx, vy)
plt.plot(x,y,'-o')
plt.axis('equal')
plt.tight_layout()
#fig.se
```
